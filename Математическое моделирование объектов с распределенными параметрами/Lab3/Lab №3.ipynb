```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10,0.1)
plt.title('График функции Cos(x)')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')
plt.grid(True)
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.arange(0.,2.5,0.1)
y1 = 1/(x + 1)
y2 = x**2 - x
y3 = np.exp(-x**2)
plt.title(r'Графики функций')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')
plt.grid(True)
plt.plot(x, y1, 'r--', x, y2, 'g^', x, y3, 'bs')
plt.legend(['1/(x - 1)','x^2 - x','e^(-x^2)'], loc=2)
plt.show()
```
