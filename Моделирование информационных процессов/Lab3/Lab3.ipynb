# Библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath
import random
```

# Постоянные переменные

```python
M = [1000,1200,1400,1600,1800]
N = 1000
T = 0
t = 0
Gamma = [0]*len(M)
Tau = [0]*len(M)
Y = []
```

# Основная программа

```python
for i in range(1,N):
    for j in range(len(M)):
        Gamma[j] = random.random()
        Tau[j] = -M[j] * math.log(Gamma[j])
    t = min(Tau)
    T += t
    Y.append(t)
M = T/N
print("Среднее время службы изделия M:", M)
```

```python
fig = plt.figure(figsize=(12, 3))
X = fig.add_subplot()
X.hist(Y,50)
X.grid()
plt.show()
```
