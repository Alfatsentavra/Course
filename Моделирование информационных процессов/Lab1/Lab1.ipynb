```python
M = 0
Pi = []
for i in range(1,301,1):
    X = np.random.uniform(0,20)
    Y = np.random.uniform(0,20)
    R = math.sqrt(X**2 + Y**2)
    if R<20:
        M += 1
    Pi.append(4*(M/i))
```

```python
X = list(range(len(Pi)))
plt.figure(figsize=(12, 3))
plt.plot(X, Pi)
plt.axhline(y = 3.14, color='red', linestyle='--')
plt.show()
```
