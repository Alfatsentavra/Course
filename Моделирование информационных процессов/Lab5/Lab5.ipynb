# Библиотеки 

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random
```

# Переменные

```python
Lambda = 100
S = 25 * 10**(-6)
l = 5 * 10**(-3)
R = 8.314
n = 6
dt = 50
Omega = (2 * Lambda * S) / (l * R * n)
```

```python
T_thau = np.zeros((50,50))
T_thau_last = np.zeros((50,50))
for i in range(4):
    T_thau_last[0][23 + i] = 1000
```

```python
fig = plt.figure(figsize=(10, 10))
plt.pcolormesh(T_thau_last, cmap='plasma', edgecolors="k", shading='flat')
```

# Программа

```python
for k in range(0,100000,dt):
    for i in range(50):
        for j in range(50):
            if i == 0 and j == 0:
                T_thau[i + 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i + 1][j])
                T_thau[i][j + 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j + 1])
                T_thau[i][j] = (1/2) * (T_thau[i + 1][j] + T_thau[i][j + 1])
            elif i == 0 and j == 49:
                T_thau[i + 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i + 1][j])
                T_thau[i][j - 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j - 1])
                T_thau[i][j] = (1/2) * (T_thau[i + 1][j] + T_thau[i][j - 1])
            elif i == 49 and j == 0:
                T_thau[i - 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i - 1][j])
                T_thau[i][j + 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j + 1])
                T_thau[i][j] = (1/2) * (T_thau[i - 1][j] + T_thau[i][j + 1])
            elif i == 49 and j == 49:
                T_thau[i - 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i - 1][j])
                T_thau[i][j - 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j - 1])
                T_thau[i][j] = (1/2) * (T_thau[i - 1][j] + T_thau[i][j - 1])
            elif i == 0:
                T_thau[i + 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i + 1][j])
                T_thau[i][j + 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j + 1])
                T_thau[i][j - 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j - 1])
                T_thau[i][j] = (1/3) * (T_thau[i + 1][j] + T_thau[i][j + 1] + T_thau[i][j - 1])
            elif i == 49:
                T_thau[i - 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i - 1][j])
                T_thau[i][j + 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j + 1])
                T_thau[i][j - 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j - 1])
                T_thau[i][j] = (1/3) * (T_thau[i - 1][j] + T_thau[i][j + 1] + T_thau[i][j - 1])
            elif j == 0:
                T_thau[i + 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i + 1][j])
                T_thau[i - 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i - 1][j])
                T_thau[i][j + 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j + 1])
                T_thau[i][j] = (1/3) * (T_thau[i + 1][j] + T_thau[i - 1][j] + T_thau[i][j + 1])
            elif j == 49:
                T_thau[i + 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i + 1][j])
                T_thau[i - 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i - 1][j])
                T_thau[i][j - 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j - 1])
                T_thau[i][j] = (1/3) * (T_thau[i + 1][j] + T_thau[i - 1][j] + T_thau[i][j - 1])
            else:
                T_thau[i + 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i + 1][j])
                T_thau[i - 1][j] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i - 1][j])
                T_thau[i][j + 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j + 1])
                T_thau[i][j - 1] = T_thau_last[i][j] - Omega * dt * (T_thau_last[i][j] - T_thau_last[i][j - 1])
                T_thau[i][j] = (1/4) * (T_thau[i + 1][j] + T_thau[i - 1][j] + T_thau[i][j + 1] + T_thau[i][j - 1])
    T_thau_last = T_thau
```
