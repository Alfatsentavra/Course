```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath

from scipy.fftpack import fft
from scipy.fftpack import ifft
```

```python
def step(x):
    N = 720
    if x < N/2:
        return 1
    else: 
        return 0
```

```python
def saw(x):
    N = 720
    if x < N/2:
        return x/(N/2)
    else:
        return (x - N/2)/(N/2)
```

```python
def delta(x):
    N = 720
    if x < N/2:
        return x/(N/2)
    else:
        return (N/2 - x) * 2/N + 1
```

# Задание №3: Амплитудный спектр

```python
def step_spec(k):
    if k > 0:
        return math.fabs(1/(math.pi*k)*(math.sin((math.pi * k)/2)))
    else:
        return 1
```

```python
def saw_spec(k):
    if k > 0:
        return math.fabs(1/(math.pi*k))
    else:
        return 0
```

```python
def delta_spec(k):
    if k % 2 == 0:
        return 0
    else:
        return 2/(math.pi*k)**2
```

```python
X = list(range(0, 720))
Y = [step(x) for x in X]
fig, axs = plt.subplots(nrows = 2, ncols = 1)
fig.set_size_inches(12, 8)
axs[0].plot(X, Y)
Yf = fft(Y)
N = len(Yf)
for i in range(N):
    Yf[i] = Yf[i]/N
Ym = []
for i in range(N):
    z = cmath.polar(Yf[i])
    y = z[0]
    Ym.append(y)
X = list(range(0, 21))
Y = [Ym[x] for x in X]
axs[1].plot(X, Y)
plt.show()
```

```python
def saw_dpf(x):
    N = 720
    if x < N/2:
        return x/(N/2)
    else:
        return ((x - N/2)/(N/2))-1
```

```python
X = list(range(0, 720))
Y = [saw_dpf(x) for x in X]
fig, axs = plt.subplots(nrows = 2, ncols = 1)
fig.set_size_inches(12, 8)
axs[0].plot(X, Y)
Yf = fft(Y)
N = len(Yf)
for i in range(N):
    Yf[i] = Yf[i]/N
Ym = []
for i in range(N):
    z = cmath.polar(Yf[i])
    y = z[0]
    Ym.append(y)
X = list(range(0, 21))
Y = [Ym[x] for x in X]
axs[1].plot(X, Y)
plt.show()
```

```python
def delta_dpf(x):
    N = 720
    if x < N/2:
        return x / (N/2)
    else:
        return (-1) * ((x - N/2)/(N/2)) + 1
```

```python
X = list(range(0, 720))
Y = [delta_dpf(x) for x in X]
fig, axs = plt.subplots(nrows = 2, ncols = 1)
fig.set_size_inches(12, 8)
axs[0].plot(X, Y)
Yf = fft(Y)
N = len(Yf)
for i in range(N):
    Yf[i] = Yf[i]/N
Ym = []
for i in range(N):
    z = cmath.polar(Yf[i])
    y = z[0]
    Ym.append(y)
X = list(range(0, 21))
Y = [Ym[x] for x in X]
axs[1].plot(X, Y)
plt.show()
```
