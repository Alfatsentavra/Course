```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath
import random
import time
```

```python
X1 = [0]*250
X2 = [0]*250
X = [0]*250
stime = 0
etime = 0
n = 5
Y = []
```

```python
for i in range(250):
    X1[i] = random.randrange(0,100,1)
    X2[i] = random.randrange(0,100,1)
```

```python
for i in range(50):
    stime = time.perf_counter()
    for j in range(n):
        X[j] = X1[j] * X2[j]
    etime = time.perf_counter()
    Y.append(etime - stime)
    n = n + 5
```

```python
X = list(range(0, 50)) 
plt.figure(figsize=(12, 3))
plt.plot(X, Y)
plt.show()
```
